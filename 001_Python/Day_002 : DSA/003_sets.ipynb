```python
##create a set
my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_set=set([1,2,3,4,5,6])
print(type(my_set))
```

```python
my_empty_set=set([1,2,3,6,5,4,5,6])
print((my_empty_set))
```

```python
##Basics Sets Operation
##Adding and Removing Elements
my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
##Remove the elements from a set
my_set.remove(3)
print(my_set)
```

```python
my_set.discard(11)
print(my_set)
```

```python
##pop method
removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
##clear all the elemnets
my_set.clear()
print(my_set)
```

```python
## Set Menmbership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

###Union
union_set=set1.union(set2)
print(union_set)

##Intersetion
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference
print(set1.difference(set2))
```

```python
set2.difference(set1)
```

```python
## Symmetric Difference
set1.symmetric_difference(set2)
```

```python
## Sets Methods
set1={1,2,3,4,5}
set2={3,4,5}

## is subset
print(set1.issubset(set2))
print(set1.issuperset(set2))
```
