### Introduction to Dictionaries

##### Dictionaries are unordered collections of items. They store data in key-value pairs.Keys must be unique and immutable(e.g,strings,numbers or tupes), while values can be of any type.

```python
##Creating Dictionaries
empty_dict=()
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name": "Barsha", "Age":24, "grade": 12}
print(student)
print(type(student))
```

```python
# Single key is alwyas used
student={"name": "Barsha", "Age":24, "name": 12}
print(student)
```

```python
## Accessing Dictionaries Elements
student={"name": "Barsha", "age":24, "grade": 'A+'}
print(student)
```

```python
## Accessing Dictionary Elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', "Not Available"))
```

```python
## Modifying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements
print(student)
```

```python
student["age"]=32  ## update value for the key
print(student)
student["address"]="Dhaka"  ## added a new key and value
print(student)
```

```python
del student['grade']  ## deleting key and value pair

print(student)
```

```python
## Dictionary methods

keys=student.keys()  ## get all the keys
print(keys)
values=student.values()  ##get all values
print(values)

items=student.items() ## get all key value pairs
print(items)
```

```python
## Shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Halima"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()  ## shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="Jannat"
print(student_copy1)
print(student)
```

```python
### Iterating Over Dictionaries
## you can use loops to iterate over dictionaries, keys,values or items

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
### Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionries
students={
    "student1":{"name":"Barsha","age":24},
    "student2":{"name":"shayra","age":34}
}
print(students)
```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary COmprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of element in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## marge 2 dictionaries into one

dic1={"a":1,"b":2}
dic2={"b":3,"c":4}
merged_dic={**dic1,**dic2}
print(merged_dic)
```
