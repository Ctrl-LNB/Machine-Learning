#### Pandas-DataFrame and Series

###### pandas is a powerful data manipulation library in python,widely used for data analysis and data cleaning. It provides two primary data stuctures. Series and DataFrame. A series is a one-dimensional array-like object,while a Dataframe is a two-dimensional,size-mutable, and potentially heterogenous tabular data structure with lebeled axes(rows and columns)

```python
import pandas as pd
```

```python
## Series

import pandas as pd
data=[1,2,3,4,5]
series= pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
## Create a Series from dicrtionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## Create a Dataframe from a dictionary oof list

data={
    'Name':['Barsha', 'Shayra', 'Mayra'],
    'Age':[24,25,28],
    'City':['Dahka','New York','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## Create a Data Frame from a list of Dictionries

data=[
    {'Name':'Barsha','Age':24,'City':'Dhaka'},
    {'Name':'Nila','Age':19,'City':'Dhaka'},
    {'Name':'Raisha','Age':27,'City':'Dhaka'},
    {'Name':'Halima','Age':22,'City':'Dhaka'}    
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
