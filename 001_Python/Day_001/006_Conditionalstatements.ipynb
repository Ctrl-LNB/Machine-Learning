```python
## else
## The else statement executes  block of code if condition in the if statemnet is False.
age=16

if age>=18:
    print("You are eligible for voting")
else:
    print("You are a minor")
```

```python
##elif
## The elif statement allowes you to multiple conditio. It stands for "else if"

age=20

if age<13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")
```

```python
## Nested Coditional Statements
## You can place one or more if, else statements inside another if, elif, else statemnet to create nested conditional statemnets.

## number eve, odd, negative

num=int(input("Enter the number"))

if num>=0:
    print(f"The number {num} is positive")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is zero or negative")
```

```python
## Practical Examples

## Determine if a year is a leap year using nested condition statement

year=int(input("Enter the year"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")

else:
    print(year, "is not a leap year")
```

```python
## Assignment
## Simple Calculator program
# Take user input

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+,-,*,/): ")

# Perform the requested operation
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
      result = num1 / num2
    else:
        result = "Error Division by zero."
else:
    result = "Invalid operation."

print("Result:", result)
```
