# 1. For Loop

```python
# for variable in range(ini, cond, inc/dec)

for i in range(1, 6, 1):
    print(i)
```

```python
for i in range(2, 20, 2):
    print(i)
```

# 2. While Loop

```python
i = 1
while i<=5:
    print(i)
    i=i+1
```

```python
n = 5
count = 1

sum = 0

while count<=n:
    sum = sum + count
    count = count + 1

print(f"Sum : {sum}")
```

# 3. Loop Control Statements

```python
# break : stops the loop

for i in range(1, 11, 1):
    if i==7:
        break
    print(i)
```

```python
#continue   : skips the value

for i in range(2,20,2):
    if i<10:
        continue
    print(i)
```

```python
for j in range(1, 10, 1):      # 1 2 3 
    if j==4:                  # skip 4
        continue              # continue : 5, 6, 7, 8, 9
    print(j)
```

```python
# pass : buffer - that does not do anything

for i in range(1,6,1):
    if i==3:
        pass
    print(i)
```
