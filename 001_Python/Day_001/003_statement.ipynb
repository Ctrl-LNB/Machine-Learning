## 1. Conditional Statements
- if
- elif
- else
- Nested if

```python
# if  : executes when the condition is true

age = 30

if age>=18:
    print("Can Vote")
```

```python
# else   : it executes when if condition is false


age = 17

if age>=18:
    print("You can Vote")
else:
    print("You can not Vote")
```

```python
# elif : more than 1 condition 

age = 25

if age>=30:
    print("You are senior citizen")
elif age >= 18:
    print("You are adult")
else:
    print("You are Youngster")
```

```python
# Nested if : condition within condition

a = 200
b = 12
c = 71

if a>b:
    if a>c:
        print(f"{a} is Greater")

elif b>c:
    print(f"{b} is Greater")
else:
    print(f"{c} is Greater")
```
